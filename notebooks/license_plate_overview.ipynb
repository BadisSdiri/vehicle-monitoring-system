```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
train_folder = "C:/Users/Badis/Desktop/vehicle-monitoring-system/datasets/license_plate_recognition/train/"
test_folder = "C:/Users/Badis/Desktop/vehicle-monitoring-system/datasets/license_plate_recognition/test/"
```

```python
def count_images(folder):
    return len([f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))])

num_train_images = count_images(train_folder)
num_test_images = count_images(test_folder)

print(f"Number of images in Training Dataset: {num_train_images}")
print(f"Number of images in Testing Dataset: {num_test_images}")
```

```python
def visualize_samples(folder, num_samples=5):
    sample_files = random.sample(os.listdir(folder), num_samples)
    plt.figure(figsize=(15, 10))
    for i, file_name in enumerate(sample_files):
        file_path = os.path.join(folder, file_name)
        img = Image.open(file_path)
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(file_name[:15] + ("..." if len(file_name) > 15 else ""))
    plt.tight_layout()
    plt.show()

print("\nVisualizing Random Samples from Training Dataset:")
visualize_samples(train_folder)

print("\nVisualizing Random Samples from Testing Dataset:")
visualize_samples(test_folder)
```

```python
def analyze_dimensions(folder):
    dimensions = []
    for file_name in os.listdir(folder):
        file_path = os.path.join(folder, file_name)
        with Image.open(file_path) as img:
            dimensions.append(img.size) 
    return dimensions

train_dimensions = analyze_dimensions(train_folder)
test_dimensions = analyze_dimensions(test_folder)
```

```python
plt.figure(figsize=(12, 6))
plt.hist([w for w, h in train_dimensions], bins=20, alpha=0.7, label="Train Widths")
plt.hist([h for w, h in train_dimensions], bins=20, alpha=0.7, label="Train Heights")
plt.hist([w for w, h in test_dimensions], bins=20, alpha=0.7, label="Test Widths")
plt.hist([h for w, h in test_dimensions], bins=20, alpha=0.7, label="Test Heights")
plt.title("Distribution of Image Dimensions")
plt.xlabel("Dimension (pixels)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
