```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
train_folder = "C:/Users/Badis/Desktop/vehicle-monitoring-system/datasets/driver_behavior/train/"
test_folder = "C:/Users/Badis/Desktop/vehicle-monitoring-system/datasets/driver_behavior/test/"
metadata_file = "C:/Users/Badis/Desktop/vehicle-monitoring-system/datasets/driver_behavior/dataset_metadata.csv"
```

```python
def check_file_integrity(folder):
    corrupted_files = []
    for class_name in os.listdir(folder):
        class_path = os.path.join(folder, class_name)
        if os.path.isdir(class_path):
            for file_name in os.listdir(class_path):
                file_path = os.path.join(class_path, file_name)
                try:
                    img = Image.open(file_path)
                    img.verify()  
                except (IOError, SyntaxError):
                    corrupted_files.append(file_path)
    return corrupted_files
```

```python
print("Checking file integrity for training data...")
corrupted_train = check_file_integrity(train_folder)
print(f"Found {len(corrupted_train)} corrupted files in training data.")
print("Checking file integrity for testing data...")
corrupted_test = check_file_integrity(test_folder)
print(f"Found {len(corrupted_test)} corrupted files in testing data.")
```

```python
def calculate_statistics(folder):
    stats = {
        "class": [],
        "num_files": [],
        "avg_width": [],
        "avg_height": [],
        "avg_file_size_kb": []
    }
    for class_name in os.listdir(folder):
        class_path = os.path.join(folder, class_name)
        if os.path.isdir(class_path):
            file_sizes = []
            widths = []
            heights = []
            num_files = 0

            for file_name in os.listdir(class_path):
                file_path = os.path.join(class_path, file_name)
                try:
                    with Image.open(file_path) as img:
                        widths.append(img.width)
                        heights.append(img.height)
                        file_sizes.append(os.path.getsize(file_path) / 1024)  # Size in KB
                        num_files += 1
                except (IOError, SyntaxError):
                    continue

            if num_files > 0:
                stats["class"].append(class_name)
                stats["num_files"].append(num_files)
                stats["avg_width"].append(sum(widths) / num_files)
                stats["avg_height"].append(sum(heights) / num_files)
                stats["avg_file_size_kb"].append(sum(file_sizes) / num_files)

    return pd.DataFrame(stats)
```

```python
print("Calculating statistics for training data...")
train_stats = calculate_statistics(train_folder)
print("Training Data Statistics:")
print(train_stats)
print("Calculating statistics for testing data...")
test_stats = calculate_statistics(test_folder)
print("Testing Data Statistics:")
print(test_stats)
```

```python
print("Saving dataset metadata to CSV...")
train_stats.to_csv(metadata_file.replace("dataset_metadata.csv", "train_metadata.csv"), index=False)
test_stats.to_csv(metadata_file.replace("dataset_metadata.csv", "test_metadata.csv"), index=False)
print("Metadata saved.")
```

```python
print("\nSummary of Training Data:")
print(train_stats.describe())
print("\nSummary of Testing Data:")
print(test_stats.describe())
```

```python
def plot_class_distribution(stats, title):
    plt.figure(figsize=(10, 6))
    plt.bar(stats["class"], stats["num_files"], color='skyblue', alpha=0.7)
    plt.title(title)
    plt.xlabel("Class")
    plt.ylabel("Number of Images")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_class_distribution(train_stats, "Training Data Class Distribution")
plot_class_distribution(test_stats, "Testing Data Class Distribution")
```

```python
def plot_avg_dimensions(stats, title):
    plt.figure(figsize=(10, 6))
    plt.bar(stats["class"], stats["avg_width"], label='Avg Width', alpha=0.7)
    plt.bar(stats["class"], stats["avg_height"], label='Avg Height', alpha=0.7)
    plt.title(title)
    plt.xlabel("Class")
    plt.ylabel("Pixels")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_avg_dimensions(train_stats, "Average Image Dimensions (Training Data)")
plot_avg_dimensions(test_stats, "Average Image Dimensions (Testing Data)")
```

```python
def plot_file_size_distribution(stats, title):
    plt.figure(figsize=(10, 6))
    plt.bar(stats["class"], stats["avg_file_size_kb"], color='lightgreen', alpha=0.7)
    plt.title(title)
    plt.xlabel("Class")
    plt.ylabel("Avg File Size (KB)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_file_size_distribution(train_stats, "Average File Size (Training Data)")
plot_file_size_distribution(test_stats, "Average File Size (Testing Data)")
```
